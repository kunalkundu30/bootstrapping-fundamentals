# Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Effect of sample size on estimation

## Samples drawn from population

```python
# Samples drawn from normal distribution

no_of_samples = 3000
result={}

population_mean = 0
population_std_dev = 1

for sample_size in range(2,no_of_samples+1):
    sample = np.random.normal(loc=population_mean,scale=population_std_dev,size=sample_size)
    sample_mean = sample.mean()
    sample_std_dev = np.sqrt(np.var(sample,ddof=1))
    result[sample_size] = [sample_mean, sample_std_dev]

means = [sample_result[0] for sample_result in result.values()]
std_devs = [sample_result[1] for sample_result in result.values()]
sample_sizes = result.keys()

print("Population mean: ", population_mean)
print("Population standard deviation: ", population_std_dev)

# Plot graphs
plt.figure(figsize=(12,6))
plt.plot(sample_sizes, means,label='sample')
plt.plot(sample_sizes,[population_mean]*len(sample_sizes),color='r',label='population')
plt.xlabel("Sample size")
plt.ylabel("Sample Mean")
plt.title("Variation of Sample mean with Sample size")
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.plot(sample_sizes, std_devs,label='sample')
plt.plot(sample_sizes,[population_std_dev]*len(sample_sizes),color='r',label='population')
plt.xlabel("Sample size")
plt.ylabel("Sample standard deviation")
plt.title("Variation of Sample standard deviation with Sample size")
plt.legend()
plt.show()
```

## Samples drawn from a big sample drawn from population

```python
# Samples drawn a large sample drawn from normal distribution

big_sample_size = 100
no_of_samples = 50
result={}

population_mean = 0
population_std_dev = 1

big_sample = np.random.normal(loc=population_mean,scale=population_std_dev,size=big_sample_size)

for sample_size in range(2,no_of_samples+1):
    sample = np.random.choice(big_sample, size=sample_size)
    sample_mean = sample.mean()
    sample_std_dev = np.sqrt(np.var(sample,ddof=1))
    result[sample_size] = [sample_mean, sample_std_dev]

means = [sample_result[0] for sample_result in result.values()]
std_devs = [sample_result[1] for sample_result in result.values()]
sample_sizes = result.keys()

big_sample_mean = big_sample.mean()
big_sample_std_dev = np.sqrt(np.var(big_sample,ddof=1))

print("Population mean: ", population_mean)
print("Population standard deviation: ", population_std_dev)

print("Big sample mean: ", big_sample_mean)
print("Big sample standard deviation: ", big_sample_std_dev)

# Plot graphs
plt.figure(figsize=(12,6))
plt.plot(sample_sizes, means,label='small samples')
plt.plot(sample_sizes,[population_mean]*len(sample_sizes),color='r',label='population')
plt.plot(sample_sizes,[big_sample_mean]*len(sample_sizes),color='g',label='big sample')
plt.xlabel("Sample size")
plt.ylabel("Sample Mean")
plt.title("Variation of Sample mean with Sample size")
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.plot(sample_sizes, std_devs,label='small samples')
plt.plot(sample_sizes,[population_std_dev]*len(sample_sizes),color='r',label='population')
plt.plot(sample_sizes,[big_sample_std_dev]*len(sample_sizes),color='g',label='big sample')
plt.xlabel("Sample size")
plt.ylabel("Sample standard deviation")
plt.title("Variation of Sample standard deviation with Sample size")
plt.show()
```

# Effect of bootstraping on estimation

## Bootstrap - Samples of fixed size drawn from population

```python
# Samples drawn from normal distribution

no_of_samples = 300
sample_size = 50
result={}

population_mean = 0
population_std_dev = 1

for sample_no in range(1,no_of_samples):
    sample = np.random.normal(loc=population_mean,scale=population_std_dev,size=sample_size)
    sample_mean = sample.mean()
    sample_std_dev = np.sqrt(np.var(sample,ddof=1))
    result[sample_no] = [sample_mean, sample_std_dev]

means = [sample_result[0] for sample_result in result.values()]
std_devs = [sample_result[1] for sample_result in result.values()]
sample_nos = result.keys()

print("Population mean: ", population_mean)
print("Population standard deviation: ", population_std_dev)

estimation_of_population_mean = np.array(means).mean()
estimation_of_population_std_dev = np.array(std_devs).mean()
print("Estimation of population mean (Mean of means of all bootstrap samples) :",estimation_of_population_mean)
print("Estimation of population std dev (Mean of standard deviations of all bootstrap samples) :",estimation_of_population_std_dev)

# Plot graphs
plt.figure(figsize=(12,6))
plt.plot(sample_nos, means,label='sample')
plt.plot(sample_nos,[population_mean]*len(sample_nos),color='r',label='population')
plt.xlabel("Sample size")
plt.ylabel("Sample Mean")
plt.title("Sample mean in each bootstrapped sample")
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.plot(sample_nos, std_devs,label='sample')
plt.plot(sample_nos,[population_std_dev]*len(sample_nos),color='r',label='population')
plt.xlabel("Sample size")
plt.ylabel("Sample standard deviation")
plt.title("Standard deviation in each bootstrapped sample")
plt.legend()
plt.show()
```
